### Preliminary Time Column notes:
- each participant was measured over a 5 minute period, in 5 second intervals

### Folder breakdown:
- 19 activities folders
    - 8 subfolders (a folder per person per activity. i.e. activity 1 has 8 subfolders, etc)
        - Each participant subfolder has 60 txt files associated with them 
        - Each text file has 125 rows per file
        - 60 text files x 5 seconds = 5 minutes

### Potential Breakdown:
- 5 seconds / 125 = .04 seconds per row
- 300 seconds for 5 minutes
- 300/.04 = 7500 rows (which is the total amount of rows per user, per activity)

```python
time_test = np.arange(0.04,300.04,.04).round(2)
```

```python
# Printing out x to confirm range works
# for x in time_test:
#     print(x)
# Works!
```

```python
# Making sure the seconds match up to the range of rows needed
len(time_test)
# 7500 is the end point for the first participant
```

```python
# Convert the array to a list to multiply it by 152 (8 participants * 19 activities = 152) for repitition
test = time_test.tolist()
```

```python
test = test*152
```

```python
# Confirming that the math checks out
len(test)
```

```python
# Turn the list into a dataframe
time_df = pd.DataFrame(test, columns=['time'])
time_df
```

```python
# Add the time column to existing dataframe
df['time'] = time_df
```
