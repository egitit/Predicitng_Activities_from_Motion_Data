```python
import pandas as pd
import numpy as np
file = '/Users/Lisa/_Sports_Activity_Capstone/df_with_times.csv'
df = pd.read_csv(file)
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
df.loc[df['time'] <= 60, 'time'] = 1
df.loc[ (df['time'] > 60) & (df['time'] <= 120), 'time' ] = 2
df.loc[ (df['time'] > 120) & (df['time'] <= 180), 'time' ] = 3
df.loc[ (df['time'] > 180) & (df['time'] <= 240), 'time' ] = 4
df.loc[ (df['time'] > 240) & (df['time'] <= 300), 'time' ] = 5
```

```python
means_df = df.groupby(['person', 'activity_number', 'time']).mean()
```

## Update the column names to reflect mean values

```python
means_df
```

```python
def update_names(df, string):
    '''
    updates the column names in my df
    
    parameters:
    df - dataframe to adjust
    string - the word that will be added on to the existing column name
    
    returns:
    a dictionary of key:values to update in the dataframe
    '''
    d = {}
    df.columns
    for x in df.columns:
        d[x] = x+string
    df.rename(columns=d, inplace=True)
    return df
```

```python
means_df = update_names(means_df, "_mean")
means_df
```
