```python
import time
import string
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, roc_curve, roc_auc_score
from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import GridSearchCV
```

```python
file = '/Users/Lisa/_Sports_Activity_Capstone/complete_data_with_activities.csv'
df = pd.read_csv(file)
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
X = df.drop('activity_number', axis=1)
y = df['activity_number']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
dt.score(X_test, y_test)
```
